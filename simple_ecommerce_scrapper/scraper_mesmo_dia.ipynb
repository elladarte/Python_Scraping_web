# Web Scraping do E-commerce Mesmo dia 

### Importando as Bibliotecas 

```python
from urllib.request import urlopen, urlretrieve
from bs4 import BeautifulSoup
import cssutils
import pandas as pd
```

### Configuracoes iniciais

```python
response = urlopen('https://petropolis.mesmodia.com.br/petropolis')
html = response.read().decode('utf-8')
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
anuncios = soup.findAll('div', class_="MuiPaper-root MuiCard-root jss32 MuiPaper-elevation4")
```

```python
for anuncio in anuncios:
    print(str(anuncio) + "\n\n")
```

```python
cards = []
for anuncio in anuncios:
    card = {}

    card['Title'] = anuncio.find("h6",{"class":"MuiTypography-root jss39 MuiTypography-body1"}).getText()
    card['Valor'] = anuncio.find("span",{"class":"jss37"}).getText()
    card['Vador Desconto'] = anuncio.find("span",{"class":"jss36"}).getText()
    card['Vendido'] = anuncio.find("div",{"class":"jss42"}).getText()

    div_style = anuncio.find("div", {"class":"MuiCardMedia-root jss34"})["style"]
    style = cssutils.parseStyle(div_style)
    url = style["background-image"]
    card['Imagem'] = url

    cards.append(card)
    
    # urllib.request.urlretrieve(url,"./output/img/" + card['Title'] + ".jpg")  

dataset = pd.DataFrame(cards)
dataset.to_csv('./output/data/dataset.csv', sep=';', index = False, encoding = 'utf-8-sig')
dataset
    
```
